**Tujuan:** Menganalisis dan menampilkan perbedaan durasi tidur berdasarkan kategori usia dan jenis kelamin.

**User:** a completely global

Masalah tidur adalah isu kesehatan yang universal, bahkan pada setiap kalangan usia. Peneliti, dokter, maupun individu dari berbagai latar belakang dapat memanfaatkan visualisasi ini sebagai informasi.

**Fungsi:** Menggambarkan tren rata-rata durasi tidur berdasarkan usia dan jenis kelamin yang dapat membantu dalam pengambilan keputusan dan yang memerlukan perhatian lebih lanjut terkait konteks yang dibahas pada visualisasi.

**Tone:** Dengan menggunakan palet warna pastel, membuat visual menjadi ringan untuk dilihat dan mudah dipahami secara visual, terlebih untuk user global.

```python
import numpy as np
import pandas as pd

dataset = pd.read_csv('/content/Sleep_Efficiency.csv', na_values=['#NAME?'])
print(dataset.head(5))
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# Membagi kategori usia
bins = [0, 10, 20, 30, 40, 50, 60, 70, 100]  # Rentang usia
labels = ['0-10', '11-20', '21-30', '31-40', '41-50', '51-60', '61-70', '71-100']  # Label kategori

# Menambahkan kolom kategori usia ke DataFrame
dataset['Age Group'] = pd.cut(dataset['Age'], bins=bins, labels=labels, right=False)

# Menghitung rata-rata durasi tidur berdasarkan kelompok usia yang telah dibagi dan jenis kelamin
avg_sleep_duration = dataset.groupby(['Gender', 'Age Group'])['Sleep duration'].mean().reset_index()

# Visualisasi rata-rata durasi tidur
plt.figure(figsize=(12, 6))
sns.barplot(x='Age Group', y='Sleep duration', hue='Gender', data=avg_sleep_duration, palette='pastel')
plt.title('Rata-rata Durasi Tidur Berdasarkan Jenis Kelamin dan Kategori Usia')
plt.ylabel('Durasi Tidur (jam)')
plt.xlabel('Kategori Usia')
plt.legend(title='Jenis Kelamin')
plt.tight_layout()
plt.show()
```

**Kesimpulan**

Variasi Durasi Tidur: Rata-rata durasi tidur bervariasi antara kelompok usia dan jenis kelamin tetapi tidak terlalu menunjukkan perbedaan yang signifikan antar kelompok usia. Juga dapat disimpilkan bahwa kelompok usia lebih muda (0-10 tahun) cenderung tidur lebih lama.

Berdasarkan Jenis Kelamin: Wanita sering melaporkan durasi tidur yang lebih lama dibandingkan pria.

Perubahan fisiologis dan masalah kesehatan pada usia lanjut dapat mempengaruhi durasi tidur, sedangkan faktor sosial dan psikologis berperan dalam perbedaan antara jenis kelamin.
